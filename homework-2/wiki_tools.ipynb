**Let's define a search_wikipedia function**

```python
def search_wikipedia(query:str):

    url = "https://en.wikipedia.org/w/api.php"
    
    headers = {"User-Agent": "ai-engineering-buildcamp/1.0 (educational project)"}

    params = {
        "action":"query",
        "format":"json",
        "list":"search",
        "srsearch":query
    }
    try:
        response = requests.get(url=url, headers=headers, params=params, timeout=5)
        response.raise_for_status()
        json = response.json()
        return json
    except requests.exceptions.RequestException as e:
        print("Request failed:", e)
        return None
    
```

**Question 2: How many of the results contain the word "capybara" (case-insensitive) in their title?**

```python
query = "capybara"
json_response = search_wikipedia(query=query)
```

```python
results_list = json_response["query"]['search']
```

```python
count_capybara = 0
for dic_result in results_list:
    try:
        title = dic_result['title'].lower()
        if "capybara" in title:
            count_capybara = count_capybara + 1
            #print(dic_result)
    except Exception as e:
        
        print(f"The result \n{dic_result}\n does not have a title",e)
        continue
```

**Let's define a get_page function**

https://en.wikipedia.org/w/index.php?title=PAGE_TITLE&action=raw

```python
def get_page(page_title:str):

    url = "https://en.wikipedia.org/w/index.php"
    
    headers = {"User-Agent": "ai-engineering-buildcamp/1.0 (educational project)"}

    params = {
        "title":page_title,
        "action":"raw"
    }
    try:
        response = requests.get(url=url, headers=headers, params=params, timeout=5)
        response.raise_for_status()
        text = response.text
        return text
    except requests.exceptions.RequestException as e:
        print("Request failed:", e)
        return None
    
```
