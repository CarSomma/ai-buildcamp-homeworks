# Converting PDFs to Markdown: Question 1

**Goal**: Extract text from all downloaded PDF books and save each one as a `.md` file in the `books_text/` directory.

We use the [`markitdown`](https://github.com/microsoft/markitdown) library which converts various document formats (PDF, DOCX, etc.) to Markdown. 

This notebook covers:
1. Creating the output directory `books_text/`
2. Initialising the `MarkItDown` converter
3. Converting every PDF in `books/` to Markdown
4. Answering **Question 1**: counting the lines in the *Think Python* book

```python
import os
from pathlib import Path
```

## Configuration

Set the output folder name and create it if it does not already exist.

```python
OUTPUT_FOLDER = "books_text"
```

```python
try:
    os.makedirs(name=OUTPUT_FOLDER, exist_ok=False)
except Exception as e:
        print(f" {OUTPUT_FOLDER}: {e}")
```

## Setup MarkItDown Converter

Instantiate the `MarkItDown` converter and define the target file extension (`.pdf`) so only PDF files are processed from the input directory.

```python
from markitdown import MarkItDown
```

```python
md = MarkItDown()
```

```python
target_format = (".pdf")
```

```python
INPUT_FOLDER = "books"
```

## Convert PDFs to Markdown

Scan the `books/` directory for `.pdf` files. For each file:
- Convert it using `MarkItDown`
- Strip the `.pdf` extension and append `.md`
- Write the resulting Markdown to `books_text/<filename>.md`

Conversion errors are caught per file so a single failure does not stop the whole batch.

```python
with os.scandir(path=INPUT_FOLDER) as entries:
    for entry in entries:
        if entry.is_file() and entry.name.endswith(target_format):
            try:
                source = entry.name
                content = md.convert(source=INPUT_FOLDER + "/" + source)
            except Exception as e:
                print(f"{entry.name}: {e}")
                continue
            filename_prefix = source.split(".", maxsplit=1)[0]
            filename = filename_prefix + ".md"
            with open(file=OUTPUT_FOLDER + "/" + filename, mode="w", encoding="utf-8") as file:
                file.write(content.markdown)
```

**Question 1**:

How many lines are in the extracted content from the "Think Python" book?

```python
FILENAME="think_python_2e.md"
with os.scandir(path=OUTPUT_FOLDER) as entries:
    for entry in entries:
        if entry.is_file() and entry.name.startswith(FILENAME):
            try:
                with open(file=OUTPUT_FOLDER + "/" + FILENAME, mode="r") as file:
                    lines = file.readlines()
            except Exception as e:
                 print(f"{entry.name}: {e}")
                 continue
            print(f"The number of lines are {len(lines)}")
```
