# Downloading Books from Allen Downey's Green Tea Press

**Goal**: Download all PDF books listed in a remote CSV file and save them locally in the `books/` directory.

This notebook covers:
1. Creating the output directory
2. Loading the book list from a remote CSV (hosted on GitHub)
3. Downloading each PDF and saving it with a cleaned filename

```python
import os
import pandas as pd
import requests
```

## Configuration

Set the name of the output folder where downloaded PDFs will be stored, and create it if it does not already exist.

```python
FOLDER_NAME = "books"
```

```python
try:
    os.makedirs(name=FOLDER_NAME, exist_ok=False)
except Exception as e:
        print(f" {FOLDER_NAME}: {e}")
```

## Step 1: Load Book List from CSV

Fetch the CSV file from the course GitHub repository. It contains book titles and their corresponding PDF download URLs. Display the first few rows to verify the data loaded correctly.

```python
# load book list from GitHub
URL_repo = "https://raw.githubusercontent.com/alexeygrigorev/ai-engineering-buildcamp-code/main/01-foundation/homework/books.csv"
books_df = pd.read_csv(URL_repo)
books_df.head(7)
```

## Step 2: Download All PDFs

Iterate over each row in the dataframe. For each book:
- Clean the title (lowercase, spaces replaced with underscores) to use as the filename
- Send an HTTP GET request to the PDF URL
- Save the raw bytes to `books/<title_cleaned>.pdf`

Any download errors are caught and printed without stopping the loop.

```python
for title, url in zip(books_df.title.values, books_df.pdf_url.values):
    title_cleaned = title.lower().replace(" ", "_")
    print(title_cleaned, url)
    try:
        response = requests.get(url=url)
        pdf_content = response.content
        with open(file=FOLDER_PATH / f"{title_cleaned}.pdf", mode="wb") as f:
            f.write(pdf_content)
    except Exception as e:
        print(f"Error downloading {title}: {e}")
```
